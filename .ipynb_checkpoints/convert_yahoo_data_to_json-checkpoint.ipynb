```python
import pandas as pd
import pymongo
import json
```

```python
stock_name = 'yahoo'
```

```python
data = pd.read_csv('{}.csv'.format(stock_name))
```

```python
data['Date'] = pd.to_datetime(data['Date'], dayfirst=True, format="%d/%m/%Y")
```

```python
data = data.set_index('Date')
```

```python
client = pymongo.MongoClient()
db = client.stock
```

```python
collection = db.yahoo
```

```python
collection.drop()
for date in data.index.values:
    post = {"Date":"{}".format(date),
           "Close":"{}".format((data.ix[date])['Close']),
           "Adj Close":"{}".format((data.ix[date]['Adj Close'])),
           "High":"{}".format((data.ix[date]['High'])),
           "Low":"{}".format((data.ix[date]['Low'])),
           "Open":"{}".format((data.ix[date]['Open'])),
           "Volume":"{}".format((data.ix[date]['Volume']))}
    collection.insert_one(post).inserted_id
```

```python
test = pd.DataFrame(list(collection.find())).set_index('Date')
```
